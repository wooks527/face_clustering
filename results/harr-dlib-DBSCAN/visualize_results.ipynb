# Visualize Face Clustering Results

## Load Clustering Results

```python
from matplotlib import pyplot as plt
import json
import cv2
```

```python
with open("clustered_results.json", "r") as f:
    cluster_info = json.load(f)
cluster_info.keys(), cluster_info[list(cluster_info.keys())[0]][:3]
```

## Visualize Results

### Original Video

- [https://www.youtube.com/watch?v=bUQj7Ng7PCs](https://www.youtube.com/watch?v=bUQj7Ng7PCs)

### Clustering Results

```python
def display_results(cluster_id, scenes):
    print(f"ID-{cluster_id}:")
    for scene_id, frame_ids in enumerate(scenes):
        print(f"Scene-{scene_id}:")
        print(f"Frame-{frame_ids[0]} to Frame-{frame_ids[-1]}:")
        rows, cols = 1, max(len(frame_ids), 5)
        fig = plt.figure(figsize=(8, 8))
        for i, frame_id in enumerate(frame_ids):
            fig.add_subplot(rows, cols, i+1)
            img_path = f"clustered_imgs/fid-{cluster_id}-{scene_id}-{frame_id}.jpg"
            img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
            plt.xticks([])
            plt.yticks([])
            plt.imshow(img)
        plt.show()
```

#### Clustered Images

- 사람이 아닌 부분과 사람인 부분을 혼동하는 경향이 있음
- 중간에 인식하지 못해 같은 구간에 나오는 사람인데 다른 구간으로 나누어진 경우가 많음
- 사람 얼굴이 아닌데 얼굴로 인식한 경우가 있음

```python
for cluster_id, scenes in cluster_info.items():
    if cluster_id != "-1":
        display_results(cluster_id, scenes)
```

#### Not-Clustered Images

- 눈으로 봤을 때는 Cluster를 만들 수 있는 경우인데 Cluster 형성이 안된 이미지가 꽤 있음

```python
for cluster_id, scenes in cluster_info.items():
    if cluster_id == "-1":
        display_results(cluster_id, scenes)
```
